```python
import pandas as pd
import numpy as np
```

Importando un archivo Excel y convirtiendo las columnas Working_Year y Remote_Working_Ratio a tipo string

```python
df = pd.read_excel("Datos.xlsx", sheet_name="Sheet1")
df["Working_Year"] = df["Working_Year"].astype("string")
df["Remote_Working_Ratio"] = df["Remote_Working_Ratio"].astype("string")
df
```

Eliminando primera columna del dataframe, se debe tener en cuenta que axis = 1 es cuando se desea eliminar una columna y axis = 0 cuando se desea eliminar una fila y el inplace = True es para eliminar permanentemente la columna en el dataframe

```python
df.drop("Id", axis=1, inplace=True)
```

Filtrar registros cuyo cargo sea Data Analyst y la experiencia sea MI usando variables

```python
Cargo_Trabajador = "Data Analyst"
Experiencia_Trabajador = "MI"
Data_Designation = df[(df["Designation"] == Cargo_Trabajador) & (df["Experience"] == Experiencia_Trabajador)]
# El siguiente cÃ³digo permite observar todas las filas y columnas del dataframe
Data_Designation.iloc[:, :]
```

Filtrar registros cuyo cargo sea Data Analyst y trabaje en una empresa L

```python
df[["Working_Year", "Designation", "Company_Size"]][(df["Designation"] == "Data Analyst") & (df["Company_Size"] == "L")]
```

Filtrar registros que contengan la palabra clave Engineer

```python
df[df["Designation"].str.contains("Engineer")]
```

Sueldos promedios por Designation

```python
df.groupby("Designation").mean().round(2)
```

Cantidad de trabajadores por Designation

```python
df["Designation"].value_counts()
```

```python
df.groupby("Designation").sum().loc[["Data Analyst", "Lead Data Engineer"]].sort_values(by="Salary_In_Rupees", ascending=False)
```

Usar la funciÃ³n agg para agrupar los registros por Designation y calcular el resultado min, max y la mediana de Salary_In_Rupees

```python
df.groupby(["Designation"]).agg({"Salary_In_Rupees": ["min", "max", "median"]}).sort_values(by="Designation")
```

Agrupar por Employment_Status y calcular el Salary_In_Rupees max y min

```python
df.groupby(["Employment_Status"]).agg({"Salary_In_Rupees": {"min", "max"}})
```

Crear una funciÃ³n lambda que multiple por 2 aquellos Salary_In_Rupees que sean mayor a 15.000.000

```python
df["FunciÃ³n"] = df.apply(lambda x: x["Salary_In_Rupees"] * 2 if x["Salary_In_Rupees"] > 15000000
else x["Salary_In_Rupees"], axis=1)
df
```
